# dict in data structure
1. dict is collection of items enclosed by {} - here item is a key:value pair

```python
# create dict
info={"name":"shiva","location":"hyderabad'","salary":15000}
print(info)
print(type(info))
```

```python
# methods or functions
print(dir(dict))
```

```python
#keys,values,items,get
info={"name":"shiva","location":"hyderad'","salaabry":15000}
print(info)
print(info.keys())
print(info.values())
print(info.items())
print(info.get("location"))
```

```python
# update()
info={"name":"shiva","location":"hyderad'","salaabry":15000}
print(info)
info.update({"name":"vikram"})
print(info)
info.update({"domin":"data engineer"})
print(info)
```

```python
#pop()
info={"name":"shiva","location":"hyderad'","salaabry":15000}
print(info)
info.pop("location")
print(info)
```

```python
# sets in data structure
1. set is collection of unorder unique element enclosed by {}
```

```python
# crate set
a={9,1,8,5,3,2,6,7,1,1,2,3,1}
print(a)
print(type(a))
```

```python
# methods or functions in set
print(dir(set))
```

```python
#add()
a={10,20,50,40,30}
print(a)
a.add(100)
print(a)
```

```python
#difference
a={1,2,3,4,5}
b={3,4,5,6,7,8}
c=b.difference(a)
print(c)
```

```python
#union
a={1,2,3,4,5}
b={3,4,5,6,7,8}
c=a.union(b)
print(c)
```

```python
#intersection
a={1,2,3,4,5}
b={3,4,5,6,7,8}
c=a.intersection(b)
print(c)
```

```python
#discard
a={10,20,30,40,50,60}
print(a)
a.discard(40)
print(a)
```
