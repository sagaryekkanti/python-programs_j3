```python
#write a python program to check if a number is even or odd
num= int(input("please enter number here:"))
if(num%2==0):
    print(num,"is even number")
else:
    print(num,"is odd number")
```

```python
#write a python program to find the greatest of two numbers using if-else
m1 = int(input("Enter first number: "))
m2= int(input("Enter second number: "))

if(m1>=m2):
    print(m1,"is largest number")
elif(m2>=m1):
    print(m2,"is largest number")
```

```python
num = int(input("Enter a number: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers")
elif num == 0:
    print("Factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial *= i
    print(num,"Factorial of  is", factorial)
```

```python
num = int(input("Enter a number: "))

count = 0
n = abs(num)     # Handle negative numbers

if n == 0:
    count = 1
else:
    while n > 0:
        n //= 10   # Remove last digit
        count += 1

print("Number of digits:", count)
```
